```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import StringIndexer
trening.columns
from pyspark.ml import Pipeline
```

```python
treningGenderIndexer = StringIndexer(inputCol="gender", outputCol="gender_cat")
treningOccupationIndexer = StringIndexer(inputCol="occupation", outputCol="occupation_cat")

# Pipeline sa St5ringIndexerima
treningPipeline = Pipeline(stages=[treningGenderIndexer, treningOccupationIndexer])

# Primena Pipeline-a na trening DF
treningIndexed = treningPipeline.fit(trening).transform(trening)
print("Trening indexed:")
treningIndexed.show()

unique_occupation_combinations = treningIndexed.select("occupation", "occupation_cat").distinct()
print("Indeksi za svako zanimanje")
unique_occupation_combinations.show()

gender_count = trening.groupBy("gender").agg(count("user_id").alias("count"))
gender_count.show()

gender_count = test.groupBy("gender").agg(count("user_id").alias("count"))
gender_count.show()

testGenderIndexer = StringIndexer(inputCol="gender", outputCol="gender_cat")
# Za Gender smemo da uradimo Encoding zato sto je broj muskaraca na treningu a i na testu veci od broja zena
# Kreiranje Pipeline-a sa StringIndexerom
testPipeline = Pipeline(stages=[testGenderIndexer])

# Primena Pipeline-a na test DF
testIndexed = testPipeline.fit(test).transform(test)
#testIndexed.show()
testIndexed= testIndexed.join(unique_occupation_combinations,"occupation")
print("Test indexed:")
testIndexed.show()

assembler = VectorAssembler(
    inputCols=['age', 'unknown',
 'Action',
 'Adventure',
 'Animation',
 'Children\'s',
 'Comedy',
 'Crime',
 'Documentary',
 'Drama',
 'Fantasy',
 'Film-Noir',
 'Horror',
 'Musical',
 'Mystery',
 'Romance',
 'Sci-Fi',
 'Thriller',
 'War',
 'Western',
 'Year',
 'gender_cat',
 'occupation_cat',],
    outputCol="features")

treningOutput = assembler.transform(treningIndexed)
testOutput = assembler.transform(testIndexed)

trening_final_data = treningOutput.select("movie_id","features",'rating')
test_final_data = testOutput.select("movie_id","features",'rating')

print("TRENING FINAL")
trening_final_data.show(truncate=False)
print("TEST FINAL")
test_final_data.show(truncate=False)
```

```python
############# LINEARNA REGRESIJA #############
lr = LinearRegression(labelCol='rating', featuresCol= 'features' )
lrModel = lr.fit(trening_final_data)
test_results = lrModel.evaluate(test_final_data)

unlabeled_data = test_final_data.select('features')
lr_predictions = lrModel.transform(test_final_data)

print("*****LINEARNA REGRESIJA*****")
print("RMSE: {}".format(test_results.rootMeanSquaredError))
print("MSE: {}".format(test_results.meanSquaredError))
print("R2: {}".format(test_results.r2))
lr_predictions.show()

############# RANDOM FOREST ###############
rf = RandomForestRegressor(featuresCol="features", labelCol="rating")
rf_model = rf.fit(trening_final_data)
rf_predictions = rf_model.transform(test_final_data)


evaluator = RegressionEvaluator(labelCol="rating", predictionCol="prediction")
rmse = evaluator.evaluate(rf_predictions, {evaluator.metricName: "rmse"})
mse = evaluator.evaluate(rf_predictions, {evaluator.metricName: "mse"})
r2 = evaluator.evaluate(rf_predictions, {evaluator.metricName: "r2"})
print()
print()
print("*****RANDOM FOREST*****")
print("RMSE:", rmse)
print("MSE:", mse)
print("R2:", r2)
rf_predictions.show()
```

*OPTIMIZACIJA HIPERPARAMETARA I KROS VALIDACIJA*

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
################# LINEARNA REGRESIJA  ###########################
lr = LinearRegression(featuresCol="features", labelCol="rating", maxIter=10)

# Define a grid of hyperparameters to search over
paramGrid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.1, 0.01]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()


evaluator = RegressionEvaluator(metricName="rmse",labelCol="rating")

#Krosvalidacija
crossval = CrossValidator(estimator=lr,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=5)


cv_model_lr = crossval.fit(trening_final_data)


predictions_lr = cv_model_lr.transform(test_final_data)

# Evaluacija modela
rmse = evaluator.evaluate(predictions_lr)
print("LINEARNA REGRESIJA")
print("RMSE: %g" % rmse)

# Optimalni hiperparametri
print("Best Model Hyperparameters:")
print("regParam: ", cv_model_lr.bestModel._java_obj.getRegParam())
print("elasticNetParam: ", cv_model_lr.bestModel._java_obj.getElasticNetParam())

############################### RANDOM FOREST #############################

rf = RandomForestRegressor(featuresCol="features", labelCol="rating")
paramGrid = (ParamGridBuilder()
             .addGrid(rf.numTrees, [5, 10, 20])
             .addGrid(rf.maxDepth, [3, 5, 9])
             .build())

evaluator = RegressionEvaluator(metricName="rmse",labelCol="rating")

#Krosvalidacija
cv = CrossValidator(estimator=rf,
                    estimatorParamMaps=paramGrid,
                    evaluator=evaluator,
                    numFolds=3)

cv_model_rf = cv.fit(trening_final_data)
bestModelRF = cv_model_rf.bestModel

predictions_rf = bestModelRF.transform(test_final_data)

#Evaluacija modela
rmse = evaluator.evaluate(predictions_rf)

#Optimalni hiperparametri
print()
print()
print("RANDOM FOREST")
print("RMSE: %g" % rmse)
print("Best Model Hyperparameters:")
print("numTrees: ", bestModelRF.getNumTrees)
print("maxDepth: ", bestModelRF.getMaxDepth())
```

#   **TREĆI DEO - Sistem preporuke baziran na kolaborativnom filtriranju**

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
```

```python
trening_path = os.path.abspath("u1.base")
test_path = os.path.abspath("u1.test")

trening = spark.read.csv(trening_path, inferSchema= True, header = False, sep = '\t')
test = spark.read.csv(test_path, inferSchema= True, header = False, sep = '\t')

trening = trening.select(func.col('_c0').alias('user_id'), func.col('_c1').alias('movie_id'), func.col('_c2').alias('rating'),func.col('_c3').alias('timestamp'))
test = test.select(func.col('_c0').alias('user_id'), func.col('_c1').alias('movie_id'), func.col('_c2').alias('rating'),func.col('_c3').alias('timestamp'))

als = ALS(maxIter=5, regParam=0.01, userCol="user_id", itemCol="movie_id", ratingCol="rating", coldStartStrategy='drop')
model = als.fit(trening)
predictions = model.transform(test)
predictions.show()
```

*OPTIMIZACIJA HIPERPARAMETARA I KROSVALIDACIJA*

```python
paramGrid = ParamGridBuilder() \
    .addGrid(als.rank, [5, 10, 30]) \
    .addGrid(als.regParam, [0.01, 0.05, 0.1]) \
    .addGrid(als.maxIter, [5, 10, 15]) \
    .build()

evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")


crossval = CrossValidator(estimator=als,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=3)

cvModel = crossval.fit(trening)


bestModel = cvModel.bestModel

# Pravimo predikcije na trening skupu podataka
train_predictions = bestModel.transform(trening)
train_predictions.show(100)

# Računamo RMSE na trening skupu podataka
train_rmse = evaluator.evaluate(train_predictions)
print("RMSE on training data:" + str(train_rmse))

predictions = bestModel.transform(test)
predictions.show(100)
rmse = evaluator.evaluate(predictions)
print("RMSE:" + str(rmse))

#RMSE je manji na trening skupu tako da vidimo da je došlo do overfittinga
```
